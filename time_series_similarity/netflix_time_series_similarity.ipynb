```python
def relation_count(series):
  if 3 not in series.value_counts().index.tolist():
    pos_count = 0
  else:
    pos_count = series.value_counts()[3]
  if 2 not in series.value_counts().index.tolist():
    neutral_count = 0
  else:
    neutral_count = series.value_counts()[2]
  if series.shape[0] == 0:
    return 0
  if series.shape[0] == neutral_count:
    return 0
  return pos_count/(series.shape[0] - neutral_count)
```

```python
from matplotlib import style
from scipy.signal import savgol_filter
plt.figure(figsize=(25,5))

netflix_analyst_update = analyst_update_dataset[analyst_update_dataset.Ticker == 'NFLX']
grouped = netflix_analyst_update.resample('M', on='Release date')
keys = list(grouped.groups.keys())
datetimes = keys[0:]
heights = grouped.agg({ 'Sentiment': [relation_count]})['Sentiment']['relation_count'].ewm(200.5).mean()
heights = np.array(heights)
sentiments = heights[0:]


import yfinance as yahooFinance
plt.figure(figsize=(25, 5))
netflix_history = yahooFinance.Ticker("NFLX")
plt.title('Monthly Netflix historical stock prices and monthly positive Analyst Update news ratio')
plt.xlabel('Date')
plt.ylabel('Historical stock prices')
history_dataset = netflix_history.history(start="2012-11-01", end='2020-01-31')
history_dataset.reset_index(inplace=True)
grouped_history = history_dataset.resample('M', on='Date')
dates = list(grouped_history.groups.keys())
mean_prices = grouped_history.agg({ 'Close': ['mean']})['Close']['mean'].ewm(30.5).mean()
mean_prices = mean_prices / np.max(mean_prices)
sentiments = sentiments / np.max(sentiments)
plt.plot(datetimes, sentiments, label='Sentiments (Analyst Update)')
plt.plot(dates, mean_prices, label = 'Prices')
plt.xlabel('Date')
plt.legend()
plt.savefig('analyst_update_graph.png')
```

```python
import scipy.stats as sp
from dtaidistance import dtw
print('Pearson', sp.pearsonr(sentiments, mean_prices)[0])
print('Spearman', sp.spearmanr(sentiments, mean_prices)[0])
print('Kendall\'s tau', sp.kendalltau(sentiments, mean_prices)[0])
mean_prices = mean_prices / np.max(mean_prices)
print('DTW', dtw.distance(sentiments, mean_prices))
```
