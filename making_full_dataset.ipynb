```python
import warnings
import numpy as np
from matplotlib import pyplot as plt
warnings.filterwarnings("ignore")
plt.style.use('seaborn-v0_8-whitegrid')
import pandas as pd
pd.options.display.max_rows = 100000
pd.options.display.max_colwidth = 100

# Loading financial news dataset
financial_news_dataset = pd.read_csv("us_equities_news_dataset.csv", sep=',')
financial_news_dataset.head(5)
```

```python
financial_news_dataset.shape
```

```python
financial_news_dataset.dtypes
```

```python
financial_news_dataset['release_date']  = pd.to_datetime(pd.to_datetime(financial_news_dataset['release_date']).dt.date)
plt.title('Financial news\' releases amount in different periods of time')
plt.xlabel('Date')
plt.ylabel('Amount of news')
plt.hist(financial_news_dataset['release_date'], bins=13)
plt.savefig('financial_distribution.pdf')
```

```python
twitter_dataset['created_at']  = pd.to_datetime(pd.to_datetime(twitter_dataset['created_at']).dt.date)
plt.title('Tweets\' releases amount in different periods of time')
plt.xlabel('Date')
plt.ylabel('Amount of tweets')
plt.hist(twitter_dataset['created_at'], bins=15)
plt.savefig('twitter_distribution.pdf')
```

```python
# The dataset with tweets about UBER company, finally, was not utilized in the work.
uber_tweets = pd.read_csv("Uber.csv", sep=',')
uber_tweets.language.value_counts()
uber_tweets_english = uber_tweets[uber_tweets.language=='en']
uber_tweets_english.head(5)
```

```python
all_financial_news_dataset = pd.DataFrame()
all_financial_news_dataset['Ticker'] = financial_news_dataset['ticker']
all_financial_news_dataset['Data source'] = 'Financial news'
all_financial_news_dataset['Headline'] = financial_news_dataset['title']
all_financial_news_dataset['Content'] = financial_news_dataset['content']
all_financial_news_dataset['Release date'] = financial_news_dataset['release_date']
all_financial_news_dataset.head(20)
```

```python
all_financial_news_dataset.shape
```

```python
all_tweets_dataset = pd.DataFrame()
all_tweets_dataset['Ticker'] = twitter_dataset['partition_1']
all_tweets_dataset['Data source'] = 'Twitter'
all_tweets_dataset['Headline'] = twitter_dataset['text']
all_tweets_dataset['Content'] = 'No content'
all_tweets_dataset['Release date'] = twitter_dataset['created_at']
all_tweets_dataset.head(20)
```

```python
all_tweets_dataset.shape
```

```python
uber_tweets_dataset = pd.DataFrame()
uber_tweets_dataset['Ticker'] = 'UBER'
uber_tweets_dataset['Data source'] = 'Twitter'
uber_tweets_dataset['Headline'] = uber_tweets_english['tweet']
uber_tweets_dataset['Content'] = 'No content'
uber_tweets_dataset['Release date'] = uber_tweets_english['date']
uber_tweets_dataset[uber_tweets_dataset.Ticker == np.nan]
```

```python
# The dataset with all financial news and tweets.
general_dataset = all_financial_news_dataset.append(all_tweets_dataset)
general_dataset = general_dataset.append(uber_tweets_dataset)
np.random.shuffle(general_dataset.values)
general_dataset.head(20)
```
